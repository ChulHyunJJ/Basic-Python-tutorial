### 1. 세 개의 양의 정수를 입력받아 그 합이 짝수이면 가장 큰 수를 출력하고, 홀수이면 가장 작은 수를 출력하는 프로그램을 작성하시오.

```
num1,num2,num3 = input("정수 3개를 입력하세요.").split()
num1 = int(num1)
num2 = int(num2)
num3 = int(num3)

sum = num1 + num2 + num3

if sum % 2 == 0 :
  print(max(num1,num2,num3))
else :
  print(min(num1,num2,num3))
```

### 2. 두 점 사이의 거리를 구하는 프로그램을 작성하시오.


*   두 점의 x, y 좌표를 각각 입력 받는다.
*   두 점 사이의 거리를 구하는 공식은 다음과 같다.
$${\sqrt{(x1-x2)^2 + (y1-y2)^2}}$$

*   제곱근은 (~)**0.5 로 표현한다.





```
def distance(x1,y1,x2,y2) :
  result = ((x1-x2)**2 + (y1-y2)**2)**0.5
  return result

print(distance(2,3,4,5))
```

### 3. 2~99 사이의 정수를 입력하면 그 수의 약수를 출력하는 프로그램을 작성하시오.

```
def calc(number) :
  for i in range(1,100) :
    if number % i == 0 :
      print(i)

num = input("정수를 입력해주세요 (2~99) :")
num = int(num)

calc(num)
```

### 4. 사각형의 둘레를 구하는 함수를 작성하고, 가로 세로의 길이가 각각 3, 6인 사각형의 둘레를 함수를 이용하여 구해 보시오.

```
def square_len(x,y) :
  return 2*x + 2*y

print(square_len(3,6))
```

### 5. 리스트로 작성한 학생 점수를 입력하면 최고점을 구하는 함수 high_score를 정의하고, 이 함수를 이용하여 아래 점수 중 최고점을 구하는 프로그램을 작성하시오.

score = [60, 72, 98, 48, 100, 86]

```
def high_score(score_input) :
  result = sorted(score_input,reverse=True)
  return result

score = []
grade = 0

while grade != 'q' :
  grade = input("학생 점수 입력 : ('q'입력시 취소)")
  if grade != 'q' :
    score.append(int(grade))

print(score)
print(high_score(score))

print("최고점은 {0} 입니다.".format(high_score(score)[0]))
```

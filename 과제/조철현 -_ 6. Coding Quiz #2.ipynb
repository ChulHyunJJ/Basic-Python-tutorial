## 입력값이 양수, 음수, 0인지 구분하는 코드를 작성하시오. (20점)

```
n = int(input("정수 입력:"))

## ~코드 작성
if n > 0 :
  print(n,"은 양수입니다.")
elif n < 0 :
  print(n,"은 음수입니다.")
elif n == 0 :
  print(n,"은 0입니다.")

```

## 삼각형과 사각형의 넓이는 구하는 함수를 작성하시오. (20점)

```
def triangle(w, h):
  ## ~코드 작성
  return 0.5*w*h

def rectangle(w, h):
  ## ~코드 작성
  return w*h

width = int(input('너비 입력:'))
height = int(input('높이 입력:'))

## ~코드 작성
print("삼각형의 넓이 :", triangle(width,height))
## ~코드 작성
print("사각형의 넓이 :", rectangle(width,height))
```

## 두 수를 입력받아 계산하는 함수들을 작성하시오. (20점)

```
def add(a, b) :
	## ~코드 작성
	return a+b

def sub(a, b) :
  ## ~코드 작성
	return a-b

def mult(a, b) :
	## ~코드 작성
	return a*b

def div(a, b) :
	## ~코드 작성
	return a/b

x = int(input('정수 x: '))
y = int(input('정수 y: '))

print('두 수의 덧셈 : %d' % add(x, y))
print('두 수의 뺄셈 : %d' % sub(x, y))
print('두 수의 곱셈 : %d' % mult(x, y))
print('두 수의 나눗셈 : %f' % div(x, y))
```

## 타자 게임 코드를 발전시켜 보세요. (20점)

```
import random
import time

# 10초 안에 최대한 많은 단어 타자치기로 변경

# 단어 리스트 w 선언과 초기화!!
w = ['cat', 'bird','dog', 'fox', 'monkey', 'lion', 'tiger', 'mouse', 'panda', 'frog', 'snake', 'wolf', 'horse']

n = 1 # 문제 번호로 사용될 변수

print('[타자 게임] 준비되면 엔터를 누르세요.')
input()

start = time.time() # 엔터 누르고 나서의 현재 시간을 start에 대입!
end_time = 0
real_time = 0
q = random.choice(w) # 리스트 w에서 임의의 단어 선택

while end_time <= 10 : # 10초가 넘어가기 전까지만 반복
  renew = real_time
  real_time = time.time() - start
  end_time += real_time - renew
  
  if end_time > 10 : 
    n -= 1 
    break

  print('남은 시간 : {0:.3}'.format(10-end_time))
  print('** 문제', n, '번 **')
  print(q)
  x = input()
      
  if q == x :
    print('통과')
    n = n + 1
    q = random.choice(w) # 리스트 w의 다른 내용으로 q에 대입

  else :
    print('삐익- 오타입니다.')
    continue
print("10초간 성공한 단어 갯수 : {0}개".format(n))
```

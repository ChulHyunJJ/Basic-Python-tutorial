## 1부터 100 사이의 홀수와 짝수의 누적합을 구하시오. (10점)

```
even_sum = 0
odd_sum = 0

for i in range(1, 101):
  i += 1
  if i % 2 == 0 :
    even_sum += i
  else :
    odd_sum += i

print('짝수의 합 : {0}'.format(even_sum))
print('홀수의 합 : {0}'.format(odd_sum))
```

## 2부터 100 사이의 소수를 구하는 코드를 작성하시오. (30점)

```
for i in range(2,101) :
  result = True
  for j in range(2,i) :
    if i % j == 0 : 
      result = False
      break
  
  if result == True :
    print(i)
```

## 가장 짧은 도시 이름을 출력하는 코드를 작성하시오. (30점)

```
cities = ['daejeon', 'seoul', 'daegu', 'busan', 'chuncheon', 'uijeongbu']
temp = {cities[0]}
short_len = len(cities[0])

# 코드 입력
for i in cities :

  if len(i) < short_len :
    short_len = len(i)
    temp.clear()
    temp.add(i)
  elif len(i) == short_len :
    temp.add(i)
  else :
    continue

print(temp)
```
